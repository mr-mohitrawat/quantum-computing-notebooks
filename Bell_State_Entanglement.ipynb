```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(2, 2)

qc.h(0)
qc.cx(0, 1)

# phi plus state
# pass

# phi minus state
qc.z(0)

# psi plus state
qc.x(1)

# psi minus state
qc.x(1)
qc.z(0)


qc.measure([0, 1], [0, 1])
display(qc.draw('mpl'))

sim = AerSimulator()
comp_qc = transpile(qc, sim)
result = sim.run(comp_qc, shots=1024).result()

count = result.get_counts()

plot_histogram(count)
```
